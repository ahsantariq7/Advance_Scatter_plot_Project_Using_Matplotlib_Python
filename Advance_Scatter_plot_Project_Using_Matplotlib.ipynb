```python
import matplotlib.pyplot as plt
import numpy as np

# Generate some random data
x = np.random.rand(100)
y = np.random.rand(100)
colors = np.random.rand(100)
sizes = 1000 * np.random.rand(100)
labels = np.random.choice(['A', 'B', 'C', 'D', 'E'], size=100)

# Create a scatter plot
fig, ax = plt.subplots()
scatter = ax.scatter(x, y, c=colors, s=sizes, alpha=0.5, cmap='viridis', marker='o', edgecolors='none', label=labels)

# Add labels and title
ax.set_xlabel('X Label')
ax.set_ylabel('Y Label')
ax.set_title('Scatter Plot')

# Add a colorbar
cbar = plt.colorbar(scatter)
cbar.set_label('Color Intensity')

# Customize tick marks and grid lines
ax.set_xticks(np.arange(0, 1.1, 0.2))
ax.set_yticks(np.arange(0, 1.1, 0.2))
ax.grid(True, linewidth=0.5, linestyle='--')

# Add a legend
handles, labels = scatter.legend_elements()
legend = ax.legend(handles, labels, loc="lower right", title="Sizes", markerscale=0.5, framealpha=0.5)


# Customize the color mapping
cmap = plt.get_cmap('viridis')
norm = plt.Normalize(vmin=np.min(colors), vmax=np.max(colors))
sm = plt.cm.ScalarMappable(cmap=cmap, norm=norm)
sm.set_array([])
cbar.set_ticks(np.linspace(0, 1, 5))
cbar.set_ticklabels(['Low', 'Low-Medium', 'Medium', 'Medium-High', 'High'])

# Save the plot to a file
plt.savefig('scatter_plot.png', dpi=300, bbox_inches='tight')

# Show the plot
plt.show()
```
